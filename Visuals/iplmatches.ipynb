VISUALIZATIONS

Matches Played per Season

```python
matches_per_season = df1['season'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
matches_per_season.plot(kind='bar', color='skyblue')
plt.title('Total Matches per Season')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

Most Successful Teams (Total Wins)

```python
team_wins = df1['winner'].value_counts()

plt.figure(figsize=(10, 6))
team_wins.plot(kind='barh', color='limegreen')
plt.title('Total Wins by Each Team')
plt.xlabel('Wins')
plt.ylabel('Team')
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```

Toss Decision Breakdown

```python
toss_decision = df1['toss_decision'].value_counts()

plt.figure(figsize=(6, 6))
toss_decision.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#FFD700', '#87CEFA'])
plt.title('Toss Decisions: Bat or Field')
plt.ylabel('')
plt.show()
```

Toss Winner = Match Winner?

```python
same_winner = (df1['toss_winner'] == df1['winner']).value_counts()

plt.figure(figsize=(6, 5))
same_winner.plot(kind='bar', color=['tomato', 'mediumseagreen'])
plt.xticks([0, 1], ['No', 'Yes'], rotation=0)
plt.title('Did Toss Winner Win the Match?')
plt.ylabel('Number of Matches')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

Top Venues by Number of Matches

```python
top_venues = df1['venue'].value_counts().head(10)

plt.figure(figsize=(10, 6))
top_venues.plot(kind='bar', color='plum')
plt.title('Top 10 Venues by Number of Matches')
plt.xlabel('Venue')
plt.ylabel('Match Count')
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
plt.show()
```

Head-to-Head Heatmap

```python
head_to_head = df1.pivot_table(index='team1', columns='team2', values='winner', aggfunc='count', fill_value=0)

plt.figure(figsize=(12, 10))
sns.heatmap(head_to_head, cmap='YlGnBu', annot=True, fmt='d')
plt.title('Head-to-Head Match Counts Between Teams')
plt.tight_layout()
plt.show()
```

Most Successful Teams in Finals 

```python
finals = df1[df1['match_type'].str.lower() == 'final']  

final_winners = finals['winner'].value_counts()

plt.figure(figsize=(10, 6))
final_winners.plot(kind='barh', color='goldenrod')
plt.title(' Most Successful Teams in Finals')
plt.xlabel('Number of Final Wins')
plt.ylabel('Team')
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```
