VISUALIZATIONS


Top Run Scorers

```python
top_scorers = df2.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_scorers.values, y=top_scorers.index, palette='Blues_r')
plt.title(' Top 10 Run Scorers')
plt.xlabel('Total Runs')
plt.ylabel('Batsman')
plt.tight_layout()
plt.show()
```

Top Wicket Takers

```python
wicket_kinds = ['bowled', 'caught', 'lbw', 'stumped', 'caught and bowled', 'hit wicket']
wickets = df2[df2['dismissal_kind'].isin(wicket_kinds)]
top_wicket_takers = wickets['bowler'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_wicket_takers.values, y=top_wicket_takers.index, palette='Purples_r')
plt.title(' Top 10 Wicket Takers')
plt.xlabel('Wickets')
plt.ylabel('Bowler')
plt.tight_layout()
plt.show()
```

Run Rate Over Time

```python
ball_by_ball = df2.merge(df1[['id', 'season']], left_on='match_id', right_on='id', how='left')

# Group by season: total runs and number of balls
season_runs = ball_by_ball.groupby('season')['total_runs'].sum()
season_balls = ball_by_ball.groupby('season').size()

run_rate = (season_runs / season_balls) * 6

plt.figure(figsize=(10, 5))
sns.lineplot(x=run_rate.index, y=run_rate.values, marker='o')
plt.title('Run Rate Over Seasons')
plt.xlabel('Season')
plt.ylabel('Run Rate (Runs per Over)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Economy Rate of Bowlers

```python
bowler_stats = df2.groupby('bowler').agg({
    'total_runs': 'sum',
    'ball': 'count'
}).rename(columns={'ball': 'balls'})

bowler_stats = bowler_stats[bowler_stats['balls'] >= 200]
bowler_stats['economy'] = (bowler_stats['total_runs'] / bowler_stats['balls']) * 6
top_economy = bowler_stats.sort_values('economy').head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_economy['economy'], y=top_economy.index, palette='Greens')
plt.title('Best Economy Rates (min 200 balls)')
plt.xlabel('Economy Rate')
plt.tight_layout()
plt.show()
```

Over-by-Over Match Progression

```python
match_id = 335982 # change this to any match ID
match_data = df2[df2['match_id'] == match_id]

overwise_runs = match_data.groupby(['inning', 'over'])['total_runs'].sum().reset_index()

# Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=overwise_runs, x='over', y='total_runs', hue='inning', marker='o', palette='Set1')
plt.title(f'ðŸ“Š Over-by-Over Progression - Match {match_id}')
plt.xlabel('Over')
plt.ylabel('Runs in Over')
plt.xticks(range(1, 21))
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Inning', labels=['1st Innings', '2nd Innings'])
plt.tight_layout()
plt.show()
```

Dismissal Types Analysis

```python
dismissal_counts = df2['dismissal_kind'].value_counts()

plt.figure(figsize=(10,5))
sns.barplot(x=dismissal_counts.values, y=dismissal_counts.index, palette='Set2')
plt.title(' Dismissal Types Frequency')
plt.xlabel('Count')
plt.ylabel('Dismissal Type')
plt.tight_layout()
plt.show()
```

Powerplay vs Death Overs Stats

```python
# Define powerplay (1â€“6) and death (16â€“20) overs
powerplay = df2[df2['over'].between(1,6)]
death = df2[df2['over'].between(16,20)]

# Calculate average runs per over
powerplay_avg = powerplay.groupby(['match_id', 'over'])['total_runs'].sum().groupby('match_id').mean().mean()
death_avg = death.groupby(['match_id', 'over'])['total_runs'].sum().groupby('match_id').mean().mean()

# Plot
plt.figure(figsize=(8, 6))
bars = plt.bar(['Powerplay (1-6)', 'Death Overs (16-20)'], [powerplay_avg, death_avg], color=['skyblue', 'salmon'])


plt.title(' Average Runs per Over: Powerplay vs Death Overs')
plt.ylabel('Runs per Over')
plt.ylim(0, max(powerplay_avg, death_avg) + 2)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```
